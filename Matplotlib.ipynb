## Bar chart

```python
plt.bar(np.arange(batsman_season.shape[0])-0.2,batsman_season['2015'],width=0.2)
plt.bar(np.arange(batsman_season.shape[0]),batsman_season['2016'],width=0.2)
plt.bar(np.arange(batsman_season.shape[0])+0.2,batsman_season['2017'],width=0.2)

plt.xticks(np.arange(batsman_season.shape[0]),batsman_season['batsman'])
plt.show()
```

```python
# stacked barchart
plt.bar(batsman_season['batsman'],batsman_season['2017'],label='2017')
plt.bar(batsman_season['batsman'],batsman_season['2016'],bottom=batsman_season['2017'],label='2016')
plt.bar(batsman_season['batsman'],batsman_season['2015'],bottom=batsman_season['2016'] + batsman_season['2017'],label='2015')
```

```python
batter = pd.read_csv('/content/batter (5).csv')
batter
```

```python
sample_df = batter.head(100).sample(25,random_state=6)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

# ploting line
plt.axhline(130,color='red')
plt.axvline(35,color='red')

# labels
plt.xlabel('average')
plt.ylabel('strike rate')
plt.title('Batsman strike rate Vs average')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(18,8))

ax[0].scatter(sample_df['avg'],sample_df['strike_rate'],color='red')
ax[1].scatter(batter['avg'],batter['strike_rate'])

# set title
ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_xlabel('average')
ax[0].set_ylabel('strike rate')

# ax[1].set_title('Avg Vs Strike Rate')
ax[1].set_xlabel('average')
ax[1].set_ylabel('strike rate')
```

## 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
z.shape

fig = plt.figure()

ax = fig.add_subplot(111,projection='3d')
p =ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
